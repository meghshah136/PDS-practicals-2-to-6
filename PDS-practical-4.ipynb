```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataframe = pd.read_csv("loan_data_set.csv")
```

```python
#Number of married customers 
dataframe['Married'].value_counts()
```

```python
#Number of Male & Female customers
dataframe['Gender'].value_counts()
```

```python
#Graph for Gender status
plt.bar(list(dataframe['Gender'].value_counts().keys()),list(dataframe['Gender'].value_counts()),color=["red","yellow"])
plt.xlabel("Persons")
plt.ylabel("Counts")
plt.title("Number of male and female customers")
plt.show()
```

```python
#Graph for marital status
plt.bar(list(dataframe['Married'].value_counts().keys()),list(dataframe['Married'].value_counts()),color=["blue","orange"])
plt.xlabel("Married or not")
plt.ylabel("Counts")
plt.title("Number of married and bechlore customers")
plt.show()
```
